#### Automated Function for stacked bar plot

```python
def plot_stacked_bars(dataframe, title_, colors=None, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_,
        color=colors 
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    
    # Labels
    plt.ylabel("Customer base (%)")
    
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=8):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectangles/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(), 1))
        
        # If value is 0 do not annotate
        if value == '0.0':
            continue
            
        ax.annotate(
            value,
            ((p.get_x() + p.get_width() / 2) * pad - 0.05, (p.get_y() + p.get_height() / 2) * pad),
            color=colour,
            size=textsize
        )
```

#### Visualization Function

```python
def dist_box_plot(dataframe, column, bins_=20, colors=["#9E2A2B", "#0b090a"], boxplot_color="#9E2A2B"):
    # Create a temporal dataframe with the data to be plotted
    fig, axs = plt.subplots(1, 2, figsize=(14, 6))

    temp = pd.DataFrame({
        "Retention": dataframe[dataframe["Churn"] == 'No'][column],
        "Churn": dataframe[dataframe["Churn"] == 'Yes'][column]
    })

    # Plot the histogram
    temp[["Retention", "Churn"]].plot(kind='hist', bins=bins_, ax=axs[0], stacked=True, color=colors)

    # X-axis label
    axs[0].set_xlabel(column)
    # Change the x-axis to plain style
    axs[0].ticklabel_format(style='plain', axis='x')
    axs[0].set_title(f'Distribution of {column} by Churn')

    # Box plot
    red_diamond = dict(markerfacecolor='r', marker='D')
    sns.boxplot(y=column, data=dataframe, x="Churn", flierprops=red_diamond, ax=axs[1], color = boxplot_color)

    axs[1].set_title(f'Boxplot of {column} by Churn')

    plt.tight_layout(pad=4)
    plt.show()
```

#### Data Split

```python
# Make a copy of our data
train_df = df.copy()

# Separate target variable from independent variables
y = df['Churn']
X = df.drop(columns=['customerID', 'Churn'])
print(X.shape)
print(y.shape)
```

- We spit the dataset so that we can simulate a real life situation by generating predictions for our test sample, without showing the predictive model these data points. This gives us the ability to see how well our model is able to generalise to new data, which is critical.

- We will use a 75-25% split between train and test respectively.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### 4.1. Logistic regression
---

#### Model Trainning

```python
logreg = LogisticRegression()

logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)  
```

#### Evaluation

```python
print('Logistic Regression')
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Precision:', precision_score(y_test, y_pred))
print('Recall:', recall_score(y_test, y_pred))
print('F1-score:', f1_score(y_test, y_pred))
print('AUC-ROC:', roc_auc_score(y_test, y_pred))
```

```python
tn, fp, fn, tp = metrics.confusion_matrix(y_test, y_pred).ravel()

print(f"True positives: {tp}")
print(f"False positives: {fp}")
print(f"True negatives: {tn}")
print(f"False negatives: {fn}\n")
```

```python
lcm = confusion_matrix(y_test, y_pred)

#Visualize the confusion matrix
sns.heatmap(lcm, annot =True, cmap = 'YlOrBr', fmt ='g')
plt.xlabel('predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

We have above the result of the performance of the logistic regression model based on the evaluated metrics. Let's interpret each metric:

1. **Accuracy:** Estimated the overall correctness of the model. In this case, it's approximately 81.4%, meaning that 81.4% of the predictions are correct.

2. **Precision:** The ratio of correctly predicted positive observations to the total predicted positives. A precision of 0.69 indicates that when the model predicts a positive outcome (churn), it is correct about 69% of the time.

3. **Recall (Sensitivity or True Positive Rate):** The ratio of correctly predicted positive observations to the all observations in actual class. A recall of 0.576 means that the model is capturing 57.6% of all actual positive instances.

4. **F1-score:** The harmonic mean of precision and recall. It's a balance between precision and recall. An F1-score of 0.628 is reasonable, indicating a good trade-off between precision and recall.

5. **AUC-ROC (Area Under the Receiver Operating Characteristic curve):** Measures the area under the curve of the true positive rate (recall) versus the false positive rate. An AUC-ROC of 0.74 is generally considered decent. It indicates a reasonable ability of the model to distinguish between classes.

Now, let's interpret the confusion matrix:

- **True Positives (TP):** 276 - Instances where the model correctly predicted churn.
- **False Positives (FP):** 124 - Instances where the model incorrectly predicted churn (predicted as churn but actually not).
- **True Negatives (TN):** 1158 - Instances where the model correctly predicted non-churn.
- **False Negatives (FN):** 203 - Instances where the model incorrectly predicted non-churn (predicted as non-churn but actually churn).

- So overall, we're able to accurately identify a good numbers of clients that do not churn, but we are only able to predict a little over half the cases where clients do churn. This in turn tells us that the current set of features are not discriminative enough to clearly distinguish between completely churners and non-churner
